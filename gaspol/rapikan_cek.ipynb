```python
import pandas as pd

# 1) Baca CSV (punyamu pakai pemisah ';')
df = pd.read_csv("/content/persentase_telepon_seluler_per_wilayah (1).csv", sep=';')

# kalau di Colab setelah upload biasanya jadi:
# df = pd.read_csv("persentase_telepon_seluler_per_wilayah (1).csv", sep=';')

# 2) Pastikan tipe data benar
df["Tahun"] = df["Tahun"].astype(int)

# 3) Bikin urutan kategori segmentasi: Perkotaan dulu, baru Perdesaan
order_seg = ["Perkotaan", "Perdesaan"]
df["Segmentasi Wilayah"] = pd.Categorical(
    df["Segmentasi Wilayah"],
    categories=order_seg,
    ordered=True
)

# 4) Sort sesuai kebutuhanmu
df_sorted = df.sort_values(
    ["Provinsi", "Segmentasi Wilayah", "Tahun"],
    ascending=[True, True, False]
).reset_index(drop=True)

# 5) Lihat hasil (contoh Aceh & Bali)
display(df_sorted[df_sorted["Provinsi"].isin(["Aceh", "Bali"])].head(20))

# 6) (opsional) Simpan hasil yang sudah rapi
df_sorted.to_csv("data_urut.csv", index=False, sep=';')
print("Saved -> data_urut.csv")
```

```python
df_sorted["Persentase"] = df_sorted["Persentase"] / 100
df_sorted["Persentase"] = df_sorted["Persentase"].apply(lambda x: f"{x:.2f}".replace(".", ","))

display(df_sorted[df_sorted["Provinsi"].isin(["Aceh", "Bali"])].head(20))
```
